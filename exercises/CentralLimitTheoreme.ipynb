# Demonstration of the central limit theoreme

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import scipy.stats as stats
import math
```

```python
def plot(means, loc, scale, bins=50):
    spread = means.max() - means.min()
    bin_size = spread/bins
    sns.histplot(x=means, weights=np.full(shape=len(means), fill_value=1/len(means)/bin_size), label="Means", bins=bins)
    x = np.linspace(loc - 3*scale, loc + 3*scale, 100)
    norm = stats.norm.pdf(x, loc, scale)
    plt.plot(x, norm, label="Normal")
    plt.legend()
```

## Binary Distribution

```python
def uniform_means(exp_size:int, n_exp:int, a=-1, b=1):
    samples = np.random.uniform(a, b, size=(n_exp, exp_size))
    means = samples.mean(axis=-1)
    return means
```

```python
a = -1
b = 1
exp_size=1
n_exp=10000
plot(uniform_means(exp_size=exp_size, n_exp=n_exp, a=a, b=b), loc=(a+b)/2, scale=(b-a)/math.sqrt(12 * exp_size), bins=50)
plt.title(f"Uniform Distribution. Experiment Size = {exp_size}")
```

## Poisson Distribution

```python
def poisson_means(exp_size:int, n_exp:int, lambda_=10):
    samples = np.random.poisson(lambda_, size=(n_exp, exp_size))
    means = samples.mean(axis=-1)
    return means
```

```python
lambda_ = 10
exp_size=2
n_exp=10000
plot(poisson_means(exp_size=exp_size, n_exp=n_exp, lambda_=lambda_), loc=lambda_, scale=math.sqrt(lambda_/exp_size), bins=20)
plt.title(f"Poisson Distribution. Experiment Size = {exp_size}")
```

## Bernoulli Distribution

```python
def bernoulli_means(exp_size:int, n_exp:int, p=0.5):
    samples = (np.random.uniform(0, 1, size=(n_exp, exp_size)) <= p).astype(float)
    means = samples.mean(axis=-1)
    return means
```

```python
p = 0.5
exp_size=1
n_exp=10000
plot(bernoulli_means(exp_size=exp_size, n_exp=n_exp, p=p), loc=p, scale=math.sqrt(p*(1-p)/exp_size), bins=50)
plt.title(f"Bernoulli Distribution. Experiment Size = {exp_size}")
```

```python
p = 0.5
exp_size=2
n_exp=10000
plot(bernoulli_means(exp_size=exp_size, n_exp=n_exp, p=p), loc=p, scale=math.sqrt(p*(1-p)/exp_size), bins=50)
plt.title(f"Bernoulli Distribution. Experiment Size = {exp_size}")
```
