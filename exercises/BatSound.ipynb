```python
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
```

```python
x = np.linspace(16, 100e3, 10000)

fig = plt.figure(figsize=(10, 5))
ax = fig.gca()
sns.lineplot(x=x, y=np.exp(-((x - 50e3) ** 2)/30e7), ax=ax, label="Спектра писка летучей мыши")
plt.grid()
plt.xlabel("Частота, Гц")
plt.ylabel("Амплитуда")
ax.axvspan(15, 20e3, alpha=0.5, color='red', label="Слышимая часть спектра")
plt.legend()
```
